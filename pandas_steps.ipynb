- *As we can see from the result of unique values of Age, there is 'None' value.
- *It is also clear from the display of the .info() method that Age column has 177 null values.
- **Thus, we need to handle these missing values.**

````python
df['Age'] = df['Age'].fillna(df['Age'].mean())
````

````python
df.Age.unique()
````

````python
df.isnull().sum()
````

- *As we can see the from the display above, the two missing values are handled properly.
- But, still the column 'Embarked' has two vlaues with missing values.
- **Let's handle them too**
  **Note**:
   - It is always possible to remove this rows too

````python
df.Embarked.unique()
````

````python
df.dropna()
````

````python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode(0))
````

````python
df.isnull().sum()
````
